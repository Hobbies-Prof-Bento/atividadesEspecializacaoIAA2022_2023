#### a) Usando um comando ifelse(), crie uma coluna chamada teen, booleana, que indica se a pessoa possui menos de 20 anos.

```r
n = nrow(student.df)
for(i in c(1:n)){
    if(i == 1){
        teen <- student.df[i,3]<20
    }
    else{
        teen <- append(teen,student.df[i,3]<20) 
    }
       
}
teen.df <-data.frame(teen = teen)
new_student.df <- cbind(student.df,teen.df)
new_student.df
```

#### a) Adicione uma quinta coluna usando as seguintes regras:
1. A 5a coluna tem o valor da coluna 2 se a coluna 1 é NA
2. A 5a coluna tem o valor da coluna 4 se a coluna 2 é NA
3. A 5a coluna contém o valor da coluna 3 em qualquer outro caso

    • O resultado deve ser:
        a b f d V5
        3 2 5 NA 5
        7 NA 2 3 3
        NA 9 5 4 9
        9 3 6 NA 6

```r
nl = nrow(mydf)
for(i in c(1:nl)){
    if(i == 1){
        if(is.na(mydf[i,1])){
            V5 <- mydf[i,2]
        }
        else{
            if(is.na(mydf[i,2])){
                V5 <- mydf[i,4]
            }
            else{
                V5 <- mydf[i,3]
            }
        }
        
    }
    else{
        if(is.na(mydf[i,1])){
            V5 <- append(V5,mydf[i,2])
        }
        else{
            if(is.na(mydf[i,2])){
                V5 <- append(V5,mydf[i,4])
            }
            else{
                V5 <- append(V5,mydf[i,3])
            }
        }
    }       
}
V5.df <-data.frame(V5 = V5)
new_mydf <- cbind(mydf,V5)
new_mydf
```

### 6. Crie uma matriz com 10 colunas contendo 100.000 números, sendo os números de 1:100000. Faça um laço for que calcula a soma de cada linha desta matriz.

```r
m2 <- matrix(c(1:100000),ncol = 10)
m2
```

```r
nl = nrow(m2)
nc = ncol(m2)
line_sum <- 0
for(i in c(1:nl)){
    sum <- 0
    for(j in c(1:nc)){
        sum <- sum + m2[i,j]
    }
    if(i == 1){
        line_sum <- sum
    }
    else{
        line_sum <- append(line_sum, sum)
    }
}
line_sum
```

```r
vector1 <- 1:10
vector2 <- c("Odd", "Loop",letters[1:8])
vector3 <- rnorm(10,sd = 10)
df1 <- data.frame(vector1, vector2, vector3, stringsAsFactors = FALSE)
df1
```

#### a)  Faça um laço genérico sobre as colunas deste data frame efetuando o seguinte cálculo:
    • Se a coluna for numérica, calcula sua média
    • Se a coluna for de texto calcula a soma dos caracteres na coluna (nchar())

```r
nl = nrow(df1)
nc = ncol(df1)

for(j in c(1:nc)){
    if(is.numeric(df1[1,j])){
        sum = 0
        for(i in c(1:nl)){
            sum = sum + df1[i,j]
        }
        print("----------")
        print(paste(" A coluna ", j, " tem a média de ", sum/nl))
    }
    else if(is.character(df1[1,j])){
        sum = 0
        for(i in c(1:nl)){
            sum = sum + nchar(df1[i,j])
        }
        print("----------")
        print(paste(" A coluna ", j, " tem ", sum, " caracteres"))
    }
}
    
```

### Funções Customizadas (UDF)

### 1. Crie um script R (chamado funcoes.R) e escreva nele as seguinte funções:

#### a) Para calcular o quadrado de um número

```r
a <- function (n){
    return(n^2)
}

a(2)
```

```r
a(8)
```

#### b)  Para receber duas matrizes e retornar a multiplicação

```r
b <- function(m1,m2){
    return(m1 %*% m2)
}
```

```r
matrix1 <-matrix(c(1:5),nrow = 1)
matrix1
```

```r
matrix2 <- matrix(c(5:1),ncol = 1)
matrix2
```

```r
b(matrix1,matrix2)
```

#### c) Para receber um data frame contendo uma coluna nome e uma coluna idade, e retornar a média das idades

```r
mydf2 <- data.frame(nome=c("João", "Maria", "José", "Ana"), idade=sample(15:30,4,replace=TRUE))
mydf2
```

```r
c <- function(df){
    return (mean(df$idade))
}
```

```r
c(mydf2)
```

#### d)  Para receber um data frame contendo uma coluna nome e uma coluna idade, e retornar uma lista com dois dados, o nome e a idade, da pessoa que contém a maior idade

```r
d <- function(df){
    lista <- list(nome = df[order(max(df$idade)),][[1]], idade = df[order(max(df$idade)),][[2]])
    return(lista)
}
```

```r
d(mydf2)
```
